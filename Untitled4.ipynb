```python
import pandas as pd
from matplotlib.pyplot import pie, axis, show
%matplotlib inline
```

```python
df = pd.read_csv('planetas.csv')

sums = df.groupby(df["diametro"])["diametro"].sum()
pie(sums, labels=sums.index);

show()
```

```python
import matplotlib.pyplot as plt
values = [60, 80, 90, 55, 10, 30]
labels = ['US', 'UK', 'India', 'Germany', 'Australia', 'South Korea']
explode = (0.2, 0, 0, 0, 0, 0)
plt.pie(values, labels= values,explode=explode,counterclock=False, shadow=True)
plt.title('Population Density Index')
plt.legend(labels,loc=3)
plt.show()
```
